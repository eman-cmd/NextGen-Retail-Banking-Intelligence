```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# 1. Create our corporate internal knowledge base documents
knowledge_documents = [
    "The standard interest rate for a Premium Savings Account is 4.25% annually, with a minimum deposit of $5,000.",
    "Customers can dispute credit card transactions within 60 days of the statement date by filling out Form 404-B.",
    "International wire transfers carry a flat fee of $35 and take 3 to 5 business days to clear completely.",
    "To close a corporate checking account, a written request signed by all registered board directors is mandatory.",
    "Personal loans offer fixed rates ranging from 5.99% to 15.49% based on the client's credit score evaluation.",
    "The daily ATM cash withdrawal limit across all standard debit cards is capped tightly at $1,000 per account."
]

# 2. Build the semantic vectorizer engine locally (Zero downloads needed!)
vectorizer = TfidfVectorizer(stop_words='english')
document_embeddings = vectorizer.fit_transform(knowledge_documents)

# 3. Define the Search Function using Cosine Similarity math
def search_knowledge_base(user_query):
    # Convert user question into the same mathematical vector structure
    query_embedding = vectorizer.transform([user_query])

    # Calculate how mathematically close the question is to each document
    similarities = cosine_similarity(query_embedding, document_embeddings).flatten()

    # Extract the highest scoring policy rule
    best_match_idx = np.argmax(similarities)
    best_score = similarities[best_match_idx]

    # Print the smart search output
    print(f"\n🔍 User Question: '{user_query}'")
    print(f"✨ Match Confidence Score: {best_score:.4f}")
    print(f"📖 Retrieved Policy Answer:\n-> {knowledge_documents[best_match_idx]}")
    print("-" * 60)

print("Search engine successfully indexed locally and ready to test!")
```

```python
# Test the localized search engine
search_knowledge_base("How do I cancel a fraudulent charge on my card?")
search_knowledge_base("What yield do I get on the premium savings option?")
search_knowledge_base("How much cash can I pull out from the machine today?")
```
